```python
# Load the Titanic dataset from the URL
df = pd.read_csv(url)

# Display a confirmation message
print("Titanic dataset loaded successfully.")
```

## Task 2: Inspect the data

Starting with displaying the first 5 rows of the dataset.

```python
# Display the first 5 rows of the dataset
print(df.head())
print(df.tail())
print(df.dtypes)
print(df.describe())
print(df.info())
```


# Task 3: Clean and preprocess the data


```python
#Create Pclass Category
df['PclassCategory'] = df['Pclass'].astype(str)

print(df[['Pclass', 'PclassCategory']].head())
```

```python
# Convert Pclass to category (string)
df['PclassCategory'] = df['Pclass'].astype(str)

# OR better (with labels)
df['PclassCategory'] = df['Pclass'].map({
    1: 'First Class',
    2: 'Second Class',
    3: 'Third Class'
})

print(df[['Pclass', 'PclassCategory']].head())
```

```python
if 'Cabin' in df.columns:
    df['Cabin'] = df['Cabin'].fillna('Unknown')
    df['CabinCategory'] = df['Cabin'].str[0]
    print(df[['Cabin', 'CabinCategory']].head())
else:
    print("Cabin column not available in this dataset")
```

```python
##Handle Missing Age

df['Age'] = df['Age'].fillna(df['Age'].mean())

print(df['Age'].head())
```

```python
##Create Family Feature

df['Family'] = df['Siblings/Spouses Aboard'] + df['Parents/Children Aboard']

print(df[['Family']].head())
```

# Task 7: Visualizations


```python
# Histogram
df['Age'].hist(bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

The histogram shows the distribution of passenger ages on the Titanic, where the x-axis represents age and the y-axis represents the number of passengers in each age group. The graph indicates that most passengers were young adults, with the highest concentration between approximately 20 and 30 years old, as shown by the tallest bars in that range. There are noticeably fewer passengers among very young children and older individuals, especially above 60 years of age. The distribution gradually decreases as age increases, forming a slightly right-skewed pattern. Overall, the chart suggests that the majority of passengers were of working age, which may have influenced patterns in survival and demographics on the ship.

```python

# Bar plot survival by class
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

The bar chart shows the survival rate of Titanic passengers by passenger class (Pclass 1, 2, and 3). The height of each bar represents the proportion of passengers who survived in each class. The graph clearly indicates that first-class passengers had the highest survival rate (around 0.63 or 63%), followed by second-class passengers (about 47%), while third-class passengers had the lowest survival rate (around 24%). This demonstrates a strong relationship between passenger class and survival chances, where higher-class passengers were more likely to survive. This pattern suggests that wealth and social status played an important role in access to rescue resources such as lifeboats.


```python
# Heatmap
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

From the heatmap, several important patterns can be observed. The Survived variable has a negative correlation with Pclass (-0.34), meaning passengers in lower classes (higher class number like 3rd class) were less likely to survive. There is also a positive correlation between Fare and Survived (0.26), suggesting that passengers who paid higher fares had better chances of survival. Additionally, Pclass and Fare have a strong negative correlation (-0.55), indicating that higher-class passengers paid more.
Another notable observation is the strong positive relationship between Family and Siblings/Spouses Aboard (0.89) and Family and Parents/Children Aboard (0.78), which is expected since the family variable is derived from these columns. Age shows weak correlations with most variables, meaning it does not strongly influence other features in this dataset.
Overall, the heatmap highlights that passenger class, fare, and family-related variables play an important role in understanding survival patterns, while some variables like age have less direct influence

```python

# Pairplot
sns.pairplot(df[['Age', 'Fare', 'Siblings/Spouses Aboard', 'Parents/Children Aboard', 'Survived']], hue='Survived')
plt.show()
```

The pair plot visualizes relationships between key numerical variables such as age, fare, and family-related features, with survival indicated by color. The distributions show that most passengers were between 20 and 40 years old, and fare values are highly skewed toward lower amounts. Scatter plots reveal that passengers who paid higher fares were more likely to survive, indicating a strong relationship between wealth and survival. Family-related variables show that most passengers traveled with few relatives, and smaller groups had better survival chances compared to larger families. Overall, the plot highlights that fare and class have a stronger influence on survival than age or family size.


# Task 8: Analyze survival rates


```python
survival_by_sex = df.groupby('Sex')['Survived'].mean()
print("Survival by sex:\n", survival_by_sex)

survival_by_class = df.groupby('Pclass')['Survived'].mean()
print("Survival by class:\n", survival_by_class)
```

The results show that survival on the Titanic was strongly influenced by both sex and passenger class. Approximately 74% of female passengers survived compared to only 19% of male passengers, indicating that women had a significantly higher chance of survival. Similarly, first-class passengers had the highest survival rate (about 63%), followed by second class (47%), while third-class passengers had the lowest survival rate (around 24%). This suggests that both gender and social class played a major role in determining survival, with women and higher-class passengers having better access to rescue opportunities.



# Task 9: Age distribution


```python

plt.figure(figsize=(12,6))
sns.boxplot(x='Pclass', y='Age', hue='Sex', data=df)
plt.title('Age Distribution by Passenger Class and Sex')
plt.show()
```

The box plot illustrates the distribution of passenger ages across different classes, with separate color groups indicating categories shown in the legend. First-class passengers tend to be older, with higher median ages and a wider range, including several elderly individuals. Second-class passengers show a moderate age distribution centered around adulthood, while third-class passengers are generally younger, with lower median ages and fewer older individuals. The presence of outliers indicates both very young children and elderly passengers in all classes. Overall, the graph highlights that age distribution varies across passenger classes, with higher-class passengers being older on average.



# Task 10: Fare distribution


```python

plt.figure(figsize=(12,6))
sns.violinplot(x='Pclass', y='Fare', hue='Survived', data=df)
plt.title('Fare Distribution by Passenger Class and Survival Status')
plt.show()
```

The plot reveals that first-class passengers paid significantly higher fares, with a wide distribution that includes very high values, while second-class fares are moderate and third-class fares are generally much lower. Additionally, within each class, the survival pattern shows that passengers who survived tend to appear more frequently at higher fare ranges, especially in first class. In contrast, lower fare passengers—particularly in third class—have a higher concentration of non-survivors.
Overall, the plot highlights a clear relationship between fare, passenger class, and survival, indicating that passengers who paid higher fares (and were typically in higher classes) had a better chance of survival, reflecting the impact of wealth and social status on survival outcomes.
